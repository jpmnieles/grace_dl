# Wing

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fn = '/home/jaynieles/dev/grace_dl/logs/wing_hid/runs/2024-07-25_00-21-03/tensorboard/version_0/delta_output_analysis.csv'
```

```python
df = pd.read_csv(fn)
```

```python
plt.figure()
plt.hist(df['delta_deg_eye_pan'], bins=100)
plt.title('Eye Pan Relative Error')
plt.grid()
plt.xlabel('Target-Predicted (deg)')
plt.ylabel('Frequency')
print('std:',df['delta_deg_eye_pan'].std())
```

```python
plt.figure()
plt.hist(df['delta_deg_eye_tilt'], bins=100)
plt.title('Eye Tilt Relative Error')
plt.xlabel('Target-Predicted (deg)')
plt.ylabel('Frequency')
plt.grid()
print('std:',df['delta_deg_eye_tilt'].std())
plt.xlim([-5,5])
```
